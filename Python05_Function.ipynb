```python
def find_max(numbers):
 if len(numbers) == 1:
  return numbers[0]
 else:
  max_of_rest = find_max(numbers[1:])
  return numbers[0] if numbers[0] > max_of_rest else max_of_rest
my_list = [3, 7, 2, 9, 1]
print(find_max(my_list))
```

```python
def count_up_to(n):
 count = 1
 while count <= n:
  yield count
  count += 1
for num in count_up_to(5):
 print(num)
```

```python
def fibonacci():
 a, b = 0, 1
 while True:
  yield a
  a, b = b, a + b
gen = fibonacci()
for _ in range(100):
 print(next(gen))
```

```python
def echo_generator():
 while True:
  received = yield
  print("Received:", received)
gen = echo_generator()
next(gen)
gen.send("Hello")
gen.send("World")
```

```python
def my_gen():
 try:
  yield 1
  yield 2
  yield 3
 finally:
  print("Generator closed")
gen = my_gen()
print(next(gen))
gen.close()
```
